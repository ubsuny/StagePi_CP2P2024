# Second Harmonic Generation (SHG)
Second Harmonic Generation (SHG) is a fascinating nonlinear optical process where two photons of the same frequency interact within a nonlinear material to create a new photon with twice the energy, and therefore, twice the frequency and half the wavelength of the original photons. This process is also known as frequency doubling.

If $P_i$ is the dipole moment of the moelcules then polarization vector is defined as the total dipole moment per unit volume. It is directly proportional to the strength of the applied electric field $\vec{E}$.
$$
\vec{P} \propto \vec{E} \\
\vec{P} = \epsilon_0 \chi \vec{E} \tag{1}
$$
where $\epsilon_0$ is the permittivity of the free space and $\chi$ is called dielectric susceptibility. The equation (1) is called linear dielectric. 

As long as the intensity of the light propagating in dielectric medium is small, all the optical parameters $\chi, \eta,$ and $\epsilon$ are constant and independent of the intensity of light. Ordinary light sources generate light of field strength of the order of $10^5$ V/m which are very small compared to the atomic field strength, and therefore cannot affect optical parameters of the medium.

Laser produces the light of field strength of order $10^7$ to $10^8$ V/m which approaches the strength of the atomic field. This intense light of laser is in a position to influence the optical parameter of the medium. Then,
$$
P  = \chi E = \chi^{(1)} E + \chi^{(2)} E^2 + \chi^{(3)} E^3 \tag{2} 
$$
$\chi^{(2)}$ and $\chi^{(3)}$ are degree of non-linearity known as non-linear susceptibility or polarizability.

A polarization oscillating at frequency $\omega$ radiates an electromagnetic waves of same frequency which propagate with same velocity as that of incident wave. Thus wave produced has same characteristics of directionality and monochromatiicty as incident wave and emitted in the same direction. This phenomenon is called SHG. 

In most crystalline materials non-linear polarizability $\chi^{(2)}$ depends on the direction of propagation, polarization of electric field and arrangement of optic axis of crystals. Since in such crystalline materials $\vec{P}$ and $\vec{E}$ are not necessarily parallel, the coefficient $\chi$ must be treated as a tensor. Then second order polarization may be represented by the relation of the form:
$$
P_i^{(2)} = \epsilon_0 \sum_{j,k} \chi_{ijk}^{(2)} E_j E_k  \tag{3}
$$

If we apply inversion operator to equation 3, we get $-P_i^{(2)} = P_i^{(2)}$ giving $\chi_{ijk}^{(2)} = 0$. This means SHG cannot occur in isotropic material. But in the case of anisotropic crystal, a uniaxial crystal, in which different material property depend upon the direction and this materials lack inversion such materials exhibit SHG.
